```python
def convert_columns(df):
    for col in df.columns:
        if col == 'ts' :
            df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')  # ts는 Int64형으로 변환
        else:
            df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)  # 나머지는 float형으로 변환
    return df  

weather_forecast_1 = convert_columns(weather_forecast_1)
weather_forecast_2 = convert_columns(weather_forecast_2)
weather_actual_1 = convert_columns(weather_actual_1)
weather_actual_2 = convert_columns(weather_actual_2)
price_real_time = convert_columns(price_real_time)
price_day_ahead = convert_columns(price_day_ahead)
market_status = convert_columns(market_status)
```

```python
def drop_missing_rows(df):
    return df.dropna(subset=['ts'])

# 각 데이터프레임에서 결측치가 있는 행을 제거
weather_forecast_1 = drop_missing_rows(weather_forecast_1)
weather_forecast_2 = drop_missing_rows(weather_forecast_2)
weather_actual_1 = drop_missing_rows(weather_actual_1)
weather_actual_2 = drop_missing_rows(weather_actual_2)
price_real_time = drop_missing_rows(price_real_time)
price_day_ahead = drop_missing_rows(price_day_ahead)
market_status = drop_missing_rows(market_status)
```

```python
def resample_to_hourly(df, time_column='ts', base_freq='15min', target_freq='h', offset='30min'):
    # ts 컬럼을 그룹화하여 평균 계산
    df = df.groupby(time_column).mean().reset_index()
    
    # UTC로 변환 후 한국 표준시로 변경
    df[time_column] = pd.to_datetime(df[time_column], unit='s', utc=True).dt.tz_convert('Asia/Seoul')
    df = df.set_index(time_column)
    
    # 15분 단위로 리샘플링하고 보간
    df = df.resample(base_freq).asfreq()
    df = df.interpolate(method='linear').ffill()
    
    # 1시간 단위로 리샘플링하여 평균 계산 (30분 오프셋 적용)
    resampled_df = df.resample(target_freq, offset=offset, label='right', closed='right').mean()
    
    # 인덱스를 재설정하고, 한국 시간대를 기준으로 타임스탬프를 초 단위로 변환
    resampled_df = resampled_df.reset_index()
    resampled_df[time_column] = (resampled_df[time_column] - pd.Timedelta(minutes=30)).astype('int64') // 10**9  # 초 단위의 정수로 변환
    
    return resampled_df

# 데이터프레임을 시간 변환된 형태로 리샘플링
weather_forecast_1 = resample_to_hourly(weather_forecast_1)
weather_forecast_2 = resample_to_hourly(weather_forecast_2)
weather_actual_1 = resample_to_hourly(weather_actual_1)
weather_actual_2 = resample_to_hourly(weather_actual_2)
price_real_time = resample_to_hourly(price_real_time)
price_day_ahead = resample_to_hourly(price_day_ahead)
market_status = resample_to_hourly(market_status)

display(weather_forecast_1)
display(weather_forecast_2)
display(weather_actual_1)
display(weather_actual_2)
display(price_real_time)
display(price_day_ahead)
display(market_status)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
import xgboost as xgb

# 1. 데이터 준비
df = merged_df.copy()  # merged_df를 사용자 제공 데이터로 가정

# Target 변수와 Feature 분리
target_column = '하루전가격(원/kWh)'
features = df.columns[df.columns != target_column]

# Feature와 Target 데이터 추출
X = df[features].values
y = df[target_column].values

# 데이터 정규화 (Min-Max Scaling)
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

# 80% 훈련, 20% 검증 데이터 분할
train_size = int(len(df) * 0.8)
X_train, X_val = X_scaled[:train_size], X_scaled[train_size:]
y_train, y_val = y_scaled[:train_size], y_scaled[train_size:]

# LSTM 입력 형태로 변환 (samples, timesteps, features)
def create_sequences(X, y, time_steps=24):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

# 각 데이터를 시계열 시퀀스로 변환
time_steps = 24  # 하루를 예측하기 위해 24시간 타임스텝 사용
X_train_seq, y_train_seq = create_sequences(X_train, y_train, time_steps)
X_val_seq, y_val_seq = create_sequences(X_val, y_val, time_steps)

# 2. LSTM 모델 정의 및 특징 추출
lstm_model = Sequential()
lstm_model.add(LSTM(50, activation='relu', return_sequences=False, input_shape=(time_steps, X_train_seq.shape[2])))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')

# 3. LSTM 모델 학습
es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

lstm_model.fit(
    X_train_seq, y_train_seq,
    validation_data=(X_val_seq, y_val_seq),
    epochs=50,
    batch_size=64,
    callbacks=[es]
)

# LSTM 특징 추출
def extract_lstm_features(model, X_seq):
    feature_extractor = Sequential(model.layers[:-1])  # 마지막 Dense 레이어 제외
    return feature_extractor.predict(X_seq)

X_train_lstm_features = extract_lstm_features(lstm_model, X_train_seq)
X_val_lstm_features = extract_lstm_features(lstm_model, X_val_seq)

# 4. XGBoost 모델 학습
# LSTM 특징과 원래 입력 특징을 합침
X_train_combined = np.hstack([X_train_lstm_features, X_train[time_steps:]])
X_val_combined = np.hstack([X_val_lstm_features, X_val[time_steps:]])
y_train_combined = y_train[time_steps:]
y_val_combined = y_val[time_steps:]

xgb_model = xgb.XGBRegressor(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.01,
    objective='reg:squarederror',
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

xgb_model.fit(X_train_combined, y_train_combined, 
              eval_set=[(X_val_combined, y_val_combined)], 
              verbose=True)

# 5. 예측 및 결과 확인
predictions = xgb_model.predict(X_val_combined)
# 예측 값 역정규화
predictions = scaler_y.inverse_transform(predictions.reshape(-1, 1))
y_val_true = scaler_y.inverse_transform(y_val_combined.reshape(-1, 1))

# 결과 출력
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 5))
plt.plot(y_val_true, label='Actual SMP')
plt.plot(predictions, label='Predicted SMP', alpha=0.7)
plt.xlabel('Time (hours)')
plt.ylabel('SMP (원/kWh)')
plt.title('Actual vs Predicted SMP (Hybrid LSTM + XGBoost)')
plt.legend()
plt.show()

# 6. e_F 계산 함수 추가
def calculate_measure(actual, forecast):
    actual = np.array(actual)
    forecast = np.array(forecast)

    positive_index = actual > 0
    negative_index = actual <= 0

    # actual이 0과 -1 사이의 값을 처리
    actual[(actual <= 0) & (actual > -1)] = -1
    
    # 양수와 음수 가격의 개수
    n1 = np.sum(positive_index) + 1e-7
    n2 = np.sum(negative_index) + 1e-7

    # e1: 양수 가격 예측 오차율
    e1 = (
        np.sum(
            np.abs(actual[positive_index] - forecast[positive_index])
            / np.abs(actual[positive_index])
        )
        / n1
    )

    # e2: 음수 가격 예측 오차율
    e2 = (
        np.sum(
            np.abs(actual[negative_index] - forecast[negative_index])
            / np.abs(actual[negative_index])
        )
        / n2
    )

    TP = np.sum((forecast > 0) & (actual > 0))
    TN = np.sum((forecast <= 0) & (actual <= 0))
    FP = np.sum((forecast > 0) & (actual <= 0))
    FN = np.sum((forecast <= 0) & (actual > 0))

    # 정확도 계산
    Accuracy = (TP + TN) / (TP + TN + FP + FN)
    print(f'Accuracy: {Accuracy}')
    print(f'e1: {e1}, e2: {e2}')

    e_F = 0.2 * e1 + 0.8 * e2 - (Accuracy - 0.95)

    return e_F

# e_F 계산 및 출력
e_F_value = calculate_measure(y_val_true, predictions)
print(f'e_F: {e_F_value}')
```
