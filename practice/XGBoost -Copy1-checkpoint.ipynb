```python
def convert_columns(df):
    for col in df.columns:
        if col == 'ts' :
            df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')  # ts는 Int64형으로 변환
        else:
            df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)  # 나머지는 float형으로 변환
    return df  

weather_forecast_1 = convert_columns(weather_forecast_1)
weather_forecast_2 = convert_columns(weather_forecast_2)

price_day_ahead = convert_columns(price_day_ahead)
```

```python
def drop_missing_rows(df):
    return df.dropna(subset=['ts'])

# 각 데이터프레임에서 결측치가 있는 행을 제거
weather_forecast_1 = drop_missing_rows(weather_forecast_1)
weather_forecast_2 = drop_missing_rows(weather_forecast_2)
price_day_ahead = drop_missing_rows(price_day_ahead)
```

```python
def resample_to_hourly(df, time_column='ts', first_freq='30min', target_freq='h', offset='30min'):
    # ts 컬럼을 그룹화하여 평균 계산
    df = df.groupby(time_column).mean().reset_index()
    
    # UTC로 변환 후 한국 표준시로 변경
    df[time_column] = pd.to_datetime(df[time_column], unit='s', utc=True).dt.tz_convert('Asia/Seoul')
    df = df.set_index(time_column)
    
    # 30분 단위로 리샘플링하고 보간
    df = df.resample(first_freq).asfreq()
    df = df.interpolate(method='linear').bfill()  # 뒤의 값을 채우는 것으로 변경
    
    # 1시간 단위로 리샘플링하여 평균 계산 (30분 오프셋 적용)
    resampled_df = df.resample(target_freq, offset=offset, label='right', closed='right').mean()
    
    # 인덱스를 재설정하고, 한국 시간대를 기준으로 타임스탬프를 초 단위로 변환
    resampled_df = resampled_df.reset_index()
    resampled_df[time_column] = (resampled_df[time_column] - pd.Timedelta(minutes=30)).astype('int64') // 10**9  # 초 단위의 정수로 변환
    
    return resampled_df

# 데이터프레임을 시간 변환된 형태로 리샘플링
weather_forecast_1 = resample_to_hourly(weather_forecast_1)
weather_forecast_2 = resample_to_hourly(weather_forecast_2)

price_day_ahead = resample_to_hourly(price_day_ahead)

display(weather_forecast_1)
display(weather_forecast_2)
display(price_day_ahead)
```

```python
merged_df = merged_df.copy()  # 슬라이스된 데이터프레임에 대한 경고 방지

# 'ts' 컬럼을 DateTime으로 변환하여 요일과 시간 정보를 구함 (한국 시간으로 변환)
merged_df['date'] = pd.to_datetime(merged_df['ts'], unit='s', utc=True).dt.tz_convert('Asia/Seoul')
merged_df['weekday'] = merged_df['date'].dt.dayofweek  # Monday=0, Sunday=6

# 각 요일에 해당하는 1, 0 변수 추가
merged_df['월요일'] = (merged_df['weekday'] == 0).astype(int)  # 1이면 월요일, 0이면 다른 요일
merged_df['평일'] = merged_df['weekday'].isin(range(1, 5)).astype(int)  # 1이면 평일 (월~금), 0이면 주말
merged_df['주말'] = merged_df['weekday'].isin([5, 6]).astype(int)  # 1이면 주말 (토~일), 0이면 평일

# 공휴일 열 추가 (특정 공휴일 데이터를 기반으로 업데이트 가능)
merged_df['공휴일'] = 0

# 불필요한 'date'와 'weekday' 컬럼 삭제
merged_df = merged_df.drop(columns=['date', 'weekday'])

merged_df
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# 설정 변수 초기화
train_window = 5  # 학습 일수
val_rmses = []  # RMSE 저장 리스트

# MinMax 스케일링 초기화
scaler = MinMaxScaler()
scaler.fit(merged_df.drop(columns=['smp_da']))  # 전체 데이터로 스케일러 학습

# 점진적 학습 및 검증 반복
for train_end in range(train_window, len(merged_df) - 1, train_window + 1):
    # 학습 및 검증 데이터 정의
    train_df = merged_df[train_end - train_window:train_end]  # train_window 만큼의 데이터 학습
    val_df = merged_df[train_end:train_end + 1]  # 다음 1일 검증

    # 피처와 타깃 나누기
    X_train = train_df.drop(columns=['smp_da'])
    y_train = train_df['smp_da']
    X_val = val_df.drop(columns=['smp_da'])
    y_val = val_df['smp_da']

    # 스케일링 적용
    X_train_scaled = scaler.transform(X_train)
    X_val_scaled = scaler.transform(X_val)

    # XGBoost 모델 학습
    xgb_model = xgb.XGBRegressor(
        objective='reg:squarederror',
        learning_rate=0.05,
        max_depth=6,
        n_estimators=100,
        subsample=0.8,
        colsample_bytree=0.8
    )
    xgb_model.fit(X_train_scaled, y_train)

    # 다음날 예측
    y_val_pred = xgb_model.predict(X_val_scaled)

    # 검증 성능 평가 (RMSE)
    val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
    val_rmses.append(val_rmse)
    print(f"Validation RMSE for day {train_end + 1}: {val_rmse}")

# 최종 평균 RMSE 출력
average_rmse = np.mean(val_rmses)
print(f"\nAverage Validation RMSE over all days: {average_rmse}")
```
