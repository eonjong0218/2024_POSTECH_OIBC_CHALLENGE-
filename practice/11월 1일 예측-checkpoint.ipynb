```python
weather_actual_1['ts'] = pd.to_datetime(weather_actual_1['ts'], unit='s')
weather_actual_2['ts'] = pd.to_datetime(weather_actual_2['ts'], unit='s')

# 빈 시간 인덱스 생성 (1시간 간격)
start_time = min(weather_actual_1['ts'].min(), weather_actual_2['ts'].min())
end_time = max(weather_actual_1['ts'].max(), weather_actual_2['ts'].max())
hourly_index = pd.date_range(start=start_time, end=end_time, freq='H')

# 원본 데이터프레임을 1시간 단위로 리샘플링하여 separate DataFrames 생성
def resample_weather_data(weather_data):
    # 빈 데이터프레임 생성
    empty_df = pd.DataFrame(index=hourly_index)

    # 빈 데이터프레임과 원본 데이터프레임 합치기
    combined_df = pd.merge(empty_df, weather_data.set_index('ts'), left_index=True, right_index=True, how='left')

    # 중복되는 시간 제거 및 시간순으로 정렬
    combined_df.sort_index(inplace=True)

    # 선형 보간으로 빈 값 채우기
    combined_df.interpolate(method='linear', inplace=True)

    # 1시간 단위로 리샘플링
    return combined_df.resample('H').first()

# 각각의 데이터프레임을 리샘플링하여 최종 결과 생성
final_weather_actual_1 = resample_weather_data(weather_actual_1)
final_weather_actual_2 = resample_weather_data(weather_actual_2)

# 결과 확인
print("Resampled weather_actual_1:")
print(final_weather_actual_1)
print("\nResampled weather_actual_2:")
print(final_weather_actual_2)
```

```python
dataframes = [
    weather_forecast_1,
    weather_forecast_2,
    weather_actual_1,
    weather_actual_2,
    price_real_time,
    market_status,
]

# price_day_ahead를 복사하여 새로운 데이터프레임 생성
merged_df = price_day_ahead.copy()

# 각 데이터프레임을 병합
for df in dataframes:
    # 중복된 ts 값 제거 (최초로 나오는 값만 남김)
    df = df.drop_duplicates(subset=['ts'])

    merged_df = pd.merge(merged_df, df, on='ts', how='left', suffixes=('', '_dup'))

    for column in df.columns:
        if column != 'ts' and column in merged_df.columns and f"{column}_dup" in merged_df.columns:
            # 겹치는 칼럼의 평균 계산
            merged_df[column] = merged_df[[column, f"{column}_dup"]].mean(axis=1)
            # 중복 칼럼 제거
            merged_df.drop(columns=[f"{column}_dup"], inplace=True)

# 결과 확인
print(merged_df.shape)  # merged_df의 행 수 확인
display(merged_df)
```

```python
import pandas as pd
import numpy as np
import random
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import tensorflow as tf
import xgboost as xgb

# 시드 고정
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)

# 1. 데이터 준비 및 전처리
features = merged_df.drop(columns=['smp_da'])  # target인 'smp_da' 컬럼을 제외한 나머지
target = merged_df['smp_da']  # 예측하고자 하는 하루전 SMP

# Feature Scaling
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# 2. 시계열 데이터를 만들기 위한 함수 정의
def create_sequences(data, target, window_size=24):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(target[i+window_size])
    return np.array(X), np.array(y)

# 시계열 데이터 생성
window_size = 24
X, y = create_sequences(features_scaled, target.values, window_size)

# 3. Train-Test Split (시간 기반 분할)
train_size = int(len(X) * 0.8)  # 80% 데이터를 학습용으로
X_train, X_val = X[:train_size], X[train_size:]
y_train, y_val = y[:train_size], y[train_size:]

# 4. LSTM 모델 정의 및 학습
lstm_model = Sequential([
    LSTM(50, input_shape=(window_size, X_train.shape[2]), return_sequences=True),
    Dropout(0.2),
    LSTM(32),
    Dropout(0.2),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mse')

# 학습
lstm_model.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val), batch_size=16)

# LSTM 모델을 통해 시계열 feature를 학습하여 임베딩된 특성 추출
lstm_features = lstm_model.predict(X)  # 모든 데이터에 대해 LSTM 임베딩을 추출

# 5. XGBoost 모델 학습
X_full = np.hstack((lstm_features, features_scaled[window_size:]))
X_train_full, X_test_full = X_full[:-24], X_full[-24:]
y_train_full = y[window_size:]

# XGBoost 시드 설정
xgb_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.05, max_depth=5, random_state=seed)
xgb_model.fit(X_train_full, y_train_full)

# 6. 예측 수행: 30일 하루전 SMP 예측
smp_predictions = xgb_model.predict(X_test_full)

# 예측 결과 확인
print("1일 예측 SMP (1시간 간격):", smp_predictions)
```

```python
import pandas as pd
import numpy as np
import random
from sklearn.preprocessing import MinMaxScaler
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras import layers, models

# 시드 고정
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)

# 1. 데이터 준비 및 전처리
features = merged_df.drop(columns=['smp_da'])  # target인 'smp_da' 컬럼을 제외한 나머지
target = merged_df['smp_da']  # 예측하고자 하는 하루전 SMP

# Feature Scaling
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# 2. 시계열 데이터를 만들기 위한 함수 정의
def create_sequences(data, target, window_size=24):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i + window_size])
        y.append(target[i + window_size])
    return np.array(X), np.array(y)

# 시계열 데이터 생성
window_size = 24
X, y = create_sequences(features_scaled, target.values, window_size)

# 3. Train-Test Split (시간 기반 분할)
train_size = int(len(X) * 0.8)  # 80% 데이터를 학습용으로
X_train, X_val = X[:train_size], X[train_size:]
y_train, y_val = y[:train_size], y[train_size:]

# 4. TCN 모델 정의
def build_tcn(input_shape):
    model = models.Sequential()
    model.add(layers.Conv1D(filters=64, kernel_size=2, padding='causal', activation='relu', input_shape=input_shape))
    model.add(layers.Conv1D(filters=64, kernel_size=2, padding='causal', activation='relu'))
    model.add(layers.Conv1D(filters=32, kernel_size=2, padding='causal', activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(1))  # 예측 값은 1개
    return model

# TCN 모델 생성
tcn_model = build_tcn((window_size, X_train.shape[2]))

# 모델 컴파일
tcn_model.compile(optimizer='adam', loss='mse')

# 5. 학습
tcn_model.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val), batch_size=16)

# 6. TCN 모델을 통해 시계열 feature를 학습하여 임베딩된 특성 추출
tcn_features = tcn_model.predict(X)  # 모든 데이터에 대해 TCN 임베딩을 추출

# 7. XGBoost 모델 학습
X_full = np.hstack((tcn_features, features_scaled[window_size:]))
X_train_full, X_test_full = X_full[:-24], X_full[-24:]
y_train_full = y[window_size:]

# XGBoost 시드 설정
xgb_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.05, max_depth=5, random_state=seed)
xgb_model.fit(X_train_full, y_train_full)

# 8. 예측 수행: 30일 하루전 SMP 예측
smp_predictions = xgb_model.predict(X_test_full)

# 예측 결과 확인
print("1일 예측 SMP (1시간 간격):", smp_predictions)
```
