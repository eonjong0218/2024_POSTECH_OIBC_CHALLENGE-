```python
def solution(bright, tile) :
    n = len(tile)
    maxi = 0
    dx, dy = [0, 1, 0, -1], [1, 0, -1, 0] # 0: 오른쪽, 1: 아래, 2: 왼쪽, 3: 위쪽
    
    # 왼쪽에서 쐈을 때
    current = 0
    for i in range(n) :
        x, y = i,0
        result = bright
        while 0 <= x < n and 0 <= y < n :
            if tile[x][y] >= 0 :
                result -= tile[x][y] 
                x, y = x + dx[current], y + dy[current]
            elif tile[x][y] == -1 :
                result = 0
                break
            elif tile[x][y] == -2 :
                current = (current + 1) % 4
                x, y = x + dx[current], y + dy[current]
        maxi = max(maxi,result)

                
    # 아래쪽에서 쐈을 때
    current = 3
    for i in range(n) :
        x, y = n-1, i
        result = bright
        while 0 <= x < n and 0 <= y < n :
            if tile[x][y] >= 0 :
                result -= tile[x][y] 
                x, y = x + dx[current], y + dy[current]
            elif tile[x][y] == -1 :
                result = 0
                break
            elif tile[x][y] == -2 :
                current = (current + 1) % 4
                x, y = x + dx[current], y + dy[current]
        maxi = max(maxi,result)

    # 위쪽에서 쐈을 때
    current = 1
    for i in range(n) :
        x, y = 0, i
        result = bright
        while 0 <= x < n and 0 <= y < n :
            if tile[x][y] >= 0 :
                result -= tile[x][y] 
                x, y = x + dx[current], y + dy[current]
            elif tile[x][y] == -1 :
                result = 0
                break
            elif tile[x][y] == -2 :
                current = (current + 1) % 4
                x, y = x + dx[current], y + dy[current]
        maxi = max(maxi,result)

    # 오른쪽에서 쐈을 때
    current = 2
    for i in range(n) :
        x, y = i, n-1
        result = bright
        while 0 <= x < n and 0 <= y < n :
            if tile[x][y] >= 0 :
                result -= tile[x][y] 
                x, y = x + dx[current], y + dy[current]
            elif tile[x][y] == -1 :
                result = 0
                break
            elif tile[x][y] == -2 :
                current = (current + 1) % 4
                x, y = x + dx[current], y + dy[current]
        maxi = max(maxi,result)

    return maxi
```
