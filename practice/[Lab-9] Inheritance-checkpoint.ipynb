### Exercise 1

1. 덧셈, 뺄셈 곱셈 나눗셈 기능을 하는 add,sub,mul,div 메소드를 가지는 class Calculator를 정의하라.
   - 모든수가 정수로 처리되며, 나눗셈을 한 이후에도 정수로 유지되도록 구현하시오.
2. print를 했을 때, value: 값이 출력되도록 구현하라. (hint: \_\_str\_\_)

```python
class Calculator:
    def __init__(self):
        self.value = 0

    def add(self,val):
        self.value += val
        
    def div(self, val_2) :
        self.value //=val_2
        
    def sub(self, val_3) :
        self.value -= val_3
    
    def mul(self, val_4) :
        self.value *= val_4
        
    def __str__(self) :
        return "value :" + str(self.value)
```

```python
c = Calculator()
c.add(50)
print(c) # Value: 50
c.div(10)
print(c) # Value: 5
c.sub(4)
print(c) # Value: 1
c.mul(2)
print(c) # Value: 2
```

### Exercise 2

Caculator 를 상속받아 제곱, 팩토리얼, 소수인지 판단하는 기능이 추가된 ScientificCalculator 를 정의하고 실험하라.

```python
class ScientificCalculator(Calculator):
    def __init__(self,value=0) :
        self.value = value
        
    def sqrt(self) :
        self.value = self.value**2
    
    def factorial(self) :
        for i in range(1,self.value) :
            self.value *= i
        
    def isprime(self) :
        for i in range(2,self.value) :
            if self.value % i == 0 :
                print(str(self.value) + " is not a prime number")
                return
        else :
            print(str(self.value) + " is a prime number")
```

```python
sc = ScientificCalculator()
sc.add(2)
print(sc) # Value: 2
sc.sqrt()
print(sc) # Value: 4
sc.factorial()
print(sc) # Value: 24

sc = ScientificCalculator(3)
sc.isprime() # 3 is a prime number

sc = ScientificCalculator(4)
sc.isprime() # 4 is not a prime number
```

아래 형상에 맞는 클래스를 생성, 상속하고 테스트해 보시오.   
($\pi=3.14$)

```python
class shape :
    def __init__(self,Colour) :
        self.Colour = Colour
        
class Rectangle(shape) :
    def __init__(self,Colour,Height,Width) :
        super().__init__(Colour)
        self.Height = Height
        self.Width = Width
        print(f'{self.Colour} rectangle created')
        
    def getArea(self) :
        print(self.Height*self.Width)
        
        
class Triangle(shape) :
    def __init__(self,Colour,Base,Height) :
        super().__init__(Colour)
        self.Base = Base
        self.Height = Height
        print(f'{self.Colour} triangle created')
        
    def getArea(self) :
        print(self.Base*self.Height//2)
        
        
class Circle(shape) :
    def __init__(self,Colour,Radius) :
        super().__init__(Colour)
        self.Radius = Radius
        print(f'{self.Colour} circle created')
        
    def getArea(self) :
        print(self.Radius**2*3.14)
    
```

```python
s1 = Rectangle('black', 2, 5)
s1.getArea()

s2 = Triangle('blue', 2, 5)
s2.getArea()

s3 = Circle('red', 5)
s3.getArea()
```
