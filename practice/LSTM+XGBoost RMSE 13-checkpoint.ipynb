```python
def convert_columns(df):
    for col in df.columns:
        if col == 'ts' :
            df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')  # ts는 Int64형으로 변환
        else:
            df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)  # 나머지는 float형으로 변환
    return df  

weather_forecast_1 = convert_columns(weather_forecast_1)
weather_forecast_2 = convert_columns(weather_forecast_2)
weather_actual_1 = convert_columns(weather_actual_1)
weather_actual_2 = convert_columns(weather_actual_2)
price_real_time = convert_columns(price_real_time)
price_day_ahead = convert_columns(price_day_ahead)
market_status = convert_columns(market_status)
```

```python
def resample_to_hourly(df, time_column='ts', base_freq='15min', target_freq='h', offset='30min'):
    # ts 컬럼을 그룹화하여 평균 계산
    df = df.groupby(time_column).mean().reset_index()
    
    # UTC로 변환 후 한국 표준시로 변경
    df[time_column] = pd.to_datetime(df[time_column], unit='s', utc=True).dt.tz_convert('Asia/Seoul')
    df = df.set_index(time_column)
    
    # 15분 단위로 리샘플링하고 보간
    df = df.resample(base_freq).asfreq()
    df = df.interpolate(method='linear').ffill()
    
    # 1시간 단위로 리샘플링하여 평균 계산 (30분 오프셋 적용)
    resampled_df = df.resample(target_freq, offset=offset, label='right', closed='right').mean()
    
    # 인덱스를 재설정하고, 한국 시간대를 기준으로 타임스탬프를 초 단위로 변환
    resampled_df = resampled_df.reset_index()
    resampled_df[time_column] = (resampled_df[time_column] - pd.Timedelta(minutes=30)).astype('int64') // 10**9  # 초 단위의 정수로 변환
    
    return resampled_df

# 데이터프레임을 시간 변환된 형태로 리샘플링
weather_forecast_1 = resample_to_hourly(weather_forecast_1)
weather_forecast_2 = resample_to_hourly(weather_forecast_2)
weather_actual_1 = resample_to_hourly(weather_actual_1)
weather_actual_2 = resample_to_hourly(weather_actual_2)
price_real_time = resample_to_hourly(price_real_time)
price_day_ahead = resample_to_hourly(price_day_ahead)
market_status = resample_to_hourly(market_status)

display(weather_forecast_1)
display(weather_forecast_2)
display(weather_actual_1)
display(weather_actual_2)
display(price_real_time)
display(price_day_ahead)
display(market_status)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Flatten, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# 예측하려는 날짜의 일부 정보를 사용하고 싶은 컬럼 목록 정의
use_cols = [
    'uv_idx', 'precip_prob', 'cloud', 'wind_dir', 'rain_prob', 'ice', 'rel_hum', 
    'dew_point', '평일', 'total_liq', 'humidity', 'snow', '주말', 'temp', 
    'ice_prob', 'vis', 'rain', 'temp_min', 'wind_speed', 'wind_spd', 'cld_cvr', 
    'ground_press', 'wind_gust_spd', 'snow_prob', 'temp_max', 'wet_bulb_temp', 
    'real_feel_temp', '월요일', 'ceiling', '공휴일', 'ts'
]

# lagged features 추가 함수 정의 (1시간, 6시간, 24시간 지연 데이터)
def create_time_lagged_features(df, target_col, lags):
    for lag in lags:
        df[f'{target_col}_lag_{lag}'] = df[target_col].shift(lag)
    df.dropna(inplace=True)
    return df

# 1. 데이터 준비
df = merged_df.copy()

# lagged features 추가
df = create_time_lagged_features(df, 'smp_da', lags=[1, 6, 24])

# 데이터 분할 (학습: 3월 ~ 10월 21일, 검증: 10월 22일)
train_df = df[(df['ts'] >= 1709218800) & (df['ts'] < 1729522800)]  # 학습 데이터 (3월 1일 ~ 10월 21일)
val_df = df[(df['ts'] >= 1729522800) & (df['ts'] <= 1729609200)]    # 검증 데이터 (10월 22일)

for col in use_cols:
    val_df.loc[:, col] = df.loc[(df['ts'] >= 1729522800) & (df['ts'] <= 1729609200), col].values

# Target과 Feature 분리
target_column = 'smp_da'
features = df.columns.difference([target_column])  # 'smp_da'만 제외하여 'ts' 포함

X_train = train_df[features].values
y_train = train_df[target_column].values

X_val = val_df[features].values
y_val = val_df[target_column].values

# 데이터 정규화 (Min-Max Scaling)
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).flatten()

X_val_scaled = scaler_X.transform(X_val)
y_val_scaled = scaler_y.transform(y_val.reshape(-1, 1)).flatten()

# 슬라이딩 윈도우 생성 (과거 1일 -> 다음날 01:00~00:00 예측)
window_size = 24  # 1일간의 데이터 (1일 * 24시간)
X_train_seq, y_train_seq = [], []

for i in range(window_size, len(X_train_scaled) - 24):
    # 입력: 이전 1일 (24시간)
    X_train_seq.append(X_train_scaled[i - window_size:i])
    
    # 타겟: 예측할 다음날 01:00 ~ 00:00 (24시간)
    y_train_seq.append(y_train_scaled[i:i + 24])

X_train_seq, y_train_seq = np.array(X_train_seq), np.array(y_train_seq)

# 검증 데이터 준비 (10월 22일 01:00 ~ 00:00 예측)
X_val_seq = np.array([X_val_scaled[-window_size:]])  # 검증일의 마지막 1일 데이터

# CNN + LSTM 모델 정의
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.2))
model.add(LSTM(50, activation='relu', return_sequences=False))
model.add(Dense(24))  # 하루 24시간 예측을 위한 출력

# 모델 컴파일
model.compile(optimizer='adam', loss='mse')

# Early Stopping과 ReduceLROnPlateau 콜백
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)

# 모델 학습
history = model.fit(X_train_seq, y_train_seq, epochs=50, batch_size=16, validation_split=0.2, 
                    callbacks=[early_stopping, reduce_lr], verbose=1)

# 검증 데이터 예측
predictions = model.predict(X_val_seq)
predictions = predictions.flatten()  # 차원을 1차원으로 변환

# 예측 값과 실제 값을 원래 스케일로 복원
predictions = scaler_y.inverse_transform(predictions.reshape(-1, 1)).flatten()
y_val_true = scaler_y.inverse_transform(y_val_scaled.reshape(-1, 1)).flatten()[-24:]  # 10월 22일 01:00 ~ 00:00 실제값

# 결과 시각화
plt.figure(figsize=(10, 5))
plt.plot(y_val_true, label="Actual", marker='o')
plt.plot(predictions, label="Predicted", marker='x')
plt.xlabel("Time Step (01:00 ~ 00:00)")
plt.ylabel("SMP")
plt.title("10월 22일 01:00 ~ 10월 23일 00:00 SMP Prediction vs Actual")
plt.legend()
plt.grid(True)
plt.show()

# RMSE 계산
rmse = np.sqrt(mean_squared_error(y_val_true, predictions))
print(f"RMSE: {rmse}")
```
